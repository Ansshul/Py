# Dealing with Null Values

   - If the Null Values don't affect the analysis we can remove them
   - If the Null values affect the analysis we can impute them.

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df
```

```python
all_data=all_data.dropna(how='all')
all_data.iloc[350:360,]                     # HERE WE CAN SEE THAT THE NULL VALUES ARE REMOVED
```

# when extracting the Month number from order date and converting it to int (not using .to_datetime here) I encountered some problem

 - FOR THAT I DID SOME SLICING AND EXTRACTING THEN CONVERTING THE MONTH TO INT AS IT RETURNED A STRING

```python
all_data=all_data[all_data['Order Date'].str[0:2]!='Or']
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

# 1. Which month has recorded the highest sales?

```python
all_data['Sales'] = all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

```python
pd.options.display.float_format = '{:.2f}'.format
results=all_data.groupby('Month',as_index=False).sum()
results['Month Name'] = pd.to_datetime(results['Month'], format='%m').dt.month_name().str.slice()
results[['Month Name','Sales']]
```

```python
months= range(1,13)
plt.figure(figsize=(12,8))
plt.ticklabel_format(style='plain')
plt.bar(results['Month Name'],results['Sales'])
plt.xticks(results['Month Name'],rotation='vertical', size=12)
#plt.yticks(results['Sales'])
plt.xlabel('Month',size=20)
plt.ylabel('Sales in USD($)',size=20)
plt.title('Sales By Month (Aggregated)',size=20)
plt.show()
```

# December recorded the highest sales (by above bar chart)

```python
all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1] + ' ' +x.split(',')[2].split(' ')[1])
all_data.head()
```

# 2. Which city has the highest sales records? 

```python
results_city=all_data.groupby('City',as_index=False).sum()
results_city[['City','Sales']]
```

```python
plt.figure(figsize=(12,8))
plt.ticklabel_format(style='plain')
plt.bar(results_city['City'], results_city['Sales'])
plt.xticks(results_city['City'], rotation='vertical', size=12)
#plt.yticks(results_city['Sales'])
plt.xlabel('City',size=20)
plt.ylabel('Sales in USD($)',size=20)
plt.title('Sales By City (Aggregated)',size=20)
plt.show()
```

# San Francisco recored the highest sales in all cities

```python
results_city[['City','Sales']].sort_values(by='Sales',ascending=False)
```

# 3.Time at which most purchases are made?

- Extracting hour an dminute from Order Date column by passing it through DateTime

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minutes'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
keys = [pair for pair, df in all_data.groupby(['Hour'])]

plt.figure(figsize=(12,8))
plt.plot(keys, all_data.groupby(['Hour']).count())
plt.xticks(keys)
plt.xlabel('Hour',size=20)
plt.ylabel('Number of Customers',size=20)
plt.title('Time at which most customers tend to buy',size=15)
plt.grid()
```

# 4. Which is the top selling product?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
quantity_ordered.sort_values(ascending=False)
```

# - It depends on the price of the commodity and also if it is a FMCG (fast moving consumer good) its quantities are high.

# - Electronic goods such as monitors,laptops and mobile devices have comparitively high price so their numbers are low(in sales)

# - Every product here can be analysed seprately :- as "27 inch FHD monitors" sells more than "Gaming monitors"


# - People here prefer Iphone over Google Phone (many more inferences can be made)

```python
products = [pair for pair, df in product_group]
plt.figure(figsize=(12,8))
plt.bar(products, quantity_ordered)
plt.xticks(products, rotation='vertical', size=12)
plt.xlabel('Product Name',size=20)
plt.ylabel('Quantities Sold',size=20)
plt.title('Top Selling Product',size=20)
plt.show()
```

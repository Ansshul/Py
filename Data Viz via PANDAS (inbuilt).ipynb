```python
df1.plot.scatter(x='A',y='B', c='C', cmap='magma')
```

# Hexagonal Bin Plot

# An alternative to scatter plot but mainly used for BIVARIATE DATA

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='magma')
```

```python
df.plot.hexbin(x='a',y='b',gridsize=15,cmap='coolwarm')
```

```python
df.plot.hexbin(x='a',y='b',gridsize=50,cmap='magma')
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25)
```
